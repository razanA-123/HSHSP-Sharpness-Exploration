```python
from StackGP import *
import numpy as np
```

```python
#Define demo function to generate data
def demoFunc(x,y):
    return x**2/y
```

```python
#Generate data
inputData=np.array([range(10),np.random.randint(1,10,10)])
response=demoFunc(inputData[0],inputData[1])
```

```python
#Generate models
models=evolve(inputData,response)
```

```python
#View model population quality
plotModels(models)
```

```python
#View best model
printGPModel(models[0])
```

```python
#Compare best model prediction to true data
plotModelResponseComparison(models[0],inputData,response)
```

```python
#Compare another model to true data
printGPModel(models[30])
plotModelResponseComparison(models[30],inputData,response)
```
